```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gaussian(mean, std, max_points=1000):
    x = np.arange(0, 5, 5/max_points)
    gaus = (1/np.sqrt(2*np.pi*(std**2)) * np.exp(-0.5 * ((x-mean)**2) / std**2))
    return x,gaus
```

```python
x,y = gaussian(mean=2,std=1, max_points=1000000)
```

```python
plt.plot(x,y)
```

```python
probs= y[::len(y)//5]
probs[-1] += 1-sum(probs)
probs
```

```python
for i in range(1000):
    print(np.random.choice(5, p=probs))
```
